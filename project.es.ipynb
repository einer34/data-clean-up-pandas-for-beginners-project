#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

ds = pd.read_csv('assets/real_estate.csv', sep=';')

casa_mas_cara = ds.loc[ds['price'].idxmax()]

lat = float(str(casa_mas_cara['latitude']).replace(',', '.'))
lon = float(str(casa_mas_cara['longitude']).replace(',', '.'))

geolocator = Nominatim(user_agent="app (einer34@hotmail.com)", timeout=10)

time.sleep(1)

ubicacion = geolocator.reverse((lat, lon), exactly_one=True)

print("La casa con dirección en", ubicacion.address, "es la más cara y su precio es de", int(casa_mas_cara['price']), "USD")

```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

ds = pd.read_csv('assets/real_estate.csv', sep=';')

casa_mas_barata = ds.loc[ds['price'].idxmin()]

lat = float(str(casa_mas_barata['latitude']).replace(',', '.'))
lon = float(str(casa_mas_barata['longitude']).replace(',', '.'))

geolocator = Nominatim(user_agent="app (einer34@hotmail.com)", timeout=10)

time.sleep(1)

ubicacion = geolocator.reverse((lat, lon), exactly_one=True)

print("La casa con dirección en", ubicacion.address, "es la más barata y su precio es de", int(casa_mas_barata['price']), "USD")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
from geopy.geocoders import Nominatim
import pandas as pd
import time

ds = pd.read_csv('assets/real_estate.csv', sep=';')
ds_filtrado = ds.dropna(subset=['surface'])

casa_mas_grande = ds_filtrado.loc[ds_filtrado['surface'].idxmax()]
casa_mas_pequena = ds_filtrado.loc[ds_filtrado['surface'].idxmin()]

geolocator = Nominatim(user_agent="app (einer34@hotmail.com)", timeout=10)

time.sleep(1)

lat_grande = float(str(casa_mas_grande['latitude']).replace(',', '.'))
lon_grande = float(str(casa_mas_grande['longitude']).replace(',', '.'))

time.sleep(1)

direccion_grande = geolocator.reverse((lat_grande, lon_grande), exactly_one=True)

lat_pequena = float(str(casa_mas_pequena['latitude']).replace(',', '.'))
lon_pequena = float(str(casa_mas_pequena['longitude']).replace(',', '.'))

time.sleep(1)
direccion_pequena = geolocator.reverse((lat_pequena, lon_pequena), exactly_one=True)

print("La casa más grande está ubicada en", direccion_grande.address,
      "y su superficie es de", int(casa_mas_grande['surface']), "metros")

print("La casa más pequeña está ubicada en", direccion_pequena.address,
      "y su superficie es de", int(casa_mas_pequena['surface']), "metros")
```

#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

`population1, population2, population3, ...`

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

filtro = ds['level5'] == "Arroyomolinos (Madrid)"
precio_promedio = ds.loc[filtro, 'price'].mean()

print("La media de precios es de:", round(precio_promedio), "USD")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python

import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

precios_arroyomolinos = ds[ds['level5'] == 'Arroyomolinos (Madrid)']['price'].dropna()

plt.figure(figsize=(10, 6))
plt.hist(precios_arroyomolinos, bins=20, color='green', edgecolor='black')
plt.title("Distribución de Precios en Arroyomolinos (Madrid)")
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
